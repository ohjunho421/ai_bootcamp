```python
columns_to_convert = ["절도(검거)", "절도(발생)", "강도(검거)", "강도(발생)", 
                      "살인(검거)", "살인(발생)", "폭력(검거)", "폭력(발생)", 
                      "소계(검거)", "소계(발생)", "강간(검거)", "강간(발생)"]

# 데이터 변환: 쉼표 제거 및 숫자로 변환
for col in columns_to_convert:
    if pdf1[col].dtype == "object":  # 문자열 데이터일 경우
        pdf1[col] = pdf1[col].str.replace(",", "").astype(float)
    else:
        pdf1[col] = pdf1[col].astype(float)

# 검거율 계산
pdf1["강간검거율"] = (pdf1["강간(검거)"] / pdf1["강간(발생)"]) * 100
pdf1["강도검거율"] = (pdf1["강도(검거)"] / pdf1["강도(발생)"]) * 100
pdf1["살인검거율"] = (pdf1["살인(검거)"] / pdf1["살인(발생)"]) * 100
pdf1["절도검거율"] = (pdf1["절도(검거)"] / pdf1["절도(발생)"]) * 100
pdf1["폭력검거율"] = (pdf1["폭력(검거)"] / pdf1["폭력(발생)"]) * 100
pdf1["검거율"] = (pdf1["소계(검거)"] / pdf1["소계(발생)"]) * 100

pdf1
```

```python
pdf1 = pdf1.drop(columns = ['강간(검거)','강도(검거)','살인(검거)','절도(검거)','폭력(검거)','소계(발생)','소계(검거)'])
pdf1
```
